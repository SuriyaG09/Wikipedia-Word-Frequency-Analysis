```python
# Import necessary libraries
import urllib.request  # Handling URL
from bs4 import BeautifulSoup  # Handling or parsing HTML files
import nltk  # Natural language toolkit

# Download the NLTK stopwords data (common words to be removed)
nltk.download('stopwords')

# Import the stopwords from NLTK
from nltk.corpus import stopwords
```

```python
# Get the content from a website
response = urllib.request.urlopen('https://en.wikipedia.org/wiki/Narendra_Modi')
html = response.read()

# Parse the HTML content using BeautifulSoup
soup = BeautifulSoup(html, 'html5lib')
```

```python
import re
```

```python

# Get the text from the HTML, stripping out any HTML tags
text = soup.get_text(strip=True)

# Define a regular expression pattern to keep only alphanumeric characters and spaces
pattern = re.compile(r'[^\w\s]')

# Remove unnecessary characters using the pattern
text = re.sub(pattern, '', text)

# Tokenize the text into words
tokens = [t for t in text.split()]
```

```python
# Get a list of English stopwords from NLTK
sr = stopwords.words('english')

# Create a copy of tokens to remove stopwords from
clean_tokens = tokens[:]
```

```python
# Iterate through tokens and remove stopwords
for token in tokens:
    if token in stopwords.words('english'):
        clean_tokens.remove(token)

# Calculate the frequency distribution of remaining words
freq = nltk.FreqDist(clean_tokens)
```

```python

# Print the word frequencies
for key, val in freq.items():
    print(str(key) + ':' + str(val))
```

```python
import matplotlib.pyplot as plt
```

```python
# Get the top 30 words and their frequencies
top_words = freq.most_common(30)

# Extract words and frequencies for the plot
words, frequencies = zip(*top_words)

# Create a bar plot for the top 30 words by frequency
plt.figure(figsize=(12, 6))
plt.bar(words, frequencies)
plt.title('Top 30 Words in the Narendra Modi Wikipedia Page')
plt.xlabel('Words')
plt.ylabel('Frequency')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

```
